```python
def separate_data_types(data):
    categorical=[]
    continuous=[]
    for column in data.columns:
        if data[column].nunique()<100:
            categorical.append(column)
        else:
            continuous.append(column)
    return categorical,continuous
```

## Plotting for Numeriacal Columns

```python
#plot data
#this function will plot the histogram, boxplot corelation heatmap for the data
#nc:numeric columns
#ctype:'h':histogram ,"b":Box plot,"c":"corelation heat map"
def plotData(data,nc,ctype):
    if ctype not in ["h","b","c"]:
        print("Invalid chart type-->{}".format(ctype))
    else:
        #corealtion heatmap
        if(ctype=="c"):
            cor=data.corr()
            cor=np.tril(cor)#fill the upper trainglw with zero
            plt.figure(figsize=(12,10))
            sns.heatmap(cor,xticklabels=nc,yticklabels=nc,vmin=-1,vmax=1,annot=True)
            plt.title("correltaion matrix")
        else:
                COLS=2;POS=1;ROWS=int(len(nc)/2)
                fig,ax=plt.subplots(ROWS,COLS,figsize=(12,10))
                for c in nc:
                    fig.add_subplot(ROWS,COLS,POS)
                    if(ctype=='h'):#histogram
                        sns.histplot(data[c],bins=20,color='red')
                    else:
                        sns.boxplot(data[c],color="orange")

                    #increaement postion by onr
                    POS+=1
    return("Success")
```

```python
def corr_col(dependent): 
    #saleprice correlation matrix
    k = len(corr_table.columns)
    cols = corr_table.nlargest(k, dependent)[dependent].index
    cm = np.corrcoef(insurance_data[cols].values.T)
    sns.set(font_scale=1.25)
    f, ax = plt.subplots(figsize=(9, 9))
    hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
    plt.show()
    d=zip(cm[0],cols)
    l_1=[]
    for _ in d:
        l_1.append(_)
    return(pd.DataFrame(l_1,columns=['Corr_value','Col_name']))
```
